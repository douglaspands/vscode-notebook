```python
import xml.etree.ElementTree as ET
import csv
from pathlib import Path

# Função para extrair dados do XML
def extrair_dados_nfe(xml_string):
    namespaces = {'nfe': 'http://www.portalfiscal.inf.br/nfe'}
    root = ET.fromstring(xml_string)

    dados = {
        'Chave de Acesso': root.find('.//nfe:infNFe', namespaces).attrib.get('Id'),
        'Data de Emissão': root.find('.//nfe:dhEmi', namespaces).text,
        'Número da NF': root.find('.//nfe:nNF', namespaces).text,
        'Emitente': root.find('.//nfe:emit/nfe:xNome', namespaces).text,
        'CNPJ Emitente': root.find('.//nfe:emit/nfe:CNPJ', namespaces).text,
        'Destinatário': root.find('.//nfe:dest/nfe:xNome', namespaces).text,
        'CPF Destinatário': root.find('.//nfe:dest/nfe:CPF', namespaces).text,
        'Valor Total': root.find('.//nfe:ICMSTot/nfe:vNF', namespaces).text,
        'Descrição do Produto': root.find('.//nfe:det/nfe:prod/nfe:xProd', namespaces).text,
        'Valor Unitário': root.find('.//nfe:det/nfe:prod/nfe:vUnCom', namespaces).text,
        'Quantidade': root.find('.//nfe:det/nfe:prod/nfe:qCom', namespaces).text,
        'Valor do ICMS': root.find('.//nfe:det/nfe:imposto/nfe:ICMS/nfe:ICMS00/nfe:vICMS', namespaces).text,
    }
    return dados

# Extrair dados
dados_nfe = extrair_dados_nfe(xml_string)

# Salvar em CSV
folder_output = Path("files/nfe/teste")
folder_output.mkdir(exist_ok=True, parents=True)
file_output = folder_output / "nfe_sp.csv"

with file_output.open(mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=dados_nfe.keys(), delimiter=';')
    writer.writeheader()
    writer.writerow(dados_nfe)

print(f"Arquivo CSV gerado com sucesso: {file_output.name}")
```
